```python
import numpy as np
import pandas as pd
from pandas import json_normalize
import json
```

```python
f = open('C:/Users/JoshuaTerc/Documents/CAT.json','r') 
data = json.load(f) 
f.close() 
```

```python
# works with check for arrayElements
df_list = []
for i in data['eventDefinitions']:
    eventtype, names, positions, required, datatype, jsondatatype, arrayElements = [], [], [], [], [], [], []
    for j in range(len(i['fields'])):
        eventtype.append(i['eventName'])
        names.append(i["fields"][j]["name"])
        positions.append(i["fields"][j]["position"])
        required.append(i["fields"][j]["required"])
        datatype.append(i["fields"][j]["dataType"])
        jsondatatype.append(i["fields"][j]["JSONDataType"])
        if len(i["fields"][j]) == 6:
            arrayElements.append(i["fields"][j]["arrayElements"])
    df = pd.DataFrame([eventtype, names, positions, required, datatype, jsondatatype, arrayElements]).T
    df.columns = ['event name', 'data element name', 'position', 'required', 'data type', 'JSON data type', 'array elements']
    df_list.append(df)
full_df = pd.concat(df_list)
full_df.head()
```

```python
# adding choices to dataframe
choices_df = json_normalize(data['choices'])
choices_df = choices_df.T.reset_index()
choices_df.columns = ['data element name', 'choices']
cat_data_definitions = pd.merge(full_df, choices_df, on='data element name', how='left')
cols = ['event name','data element name','position','required','data type','JSON data type','choices','array elements']
cat_data_definitions = cat_data_definitions[cols]
cat_data_definitions.to_csv('cat_data_definitions.csv', index=False)
cat_data_definitions.head()
```
